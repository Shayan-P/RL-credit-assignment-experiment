```python
import gymnasium as gym
env_test = gym.make('CartPole-v1', render_mode='human')
obs, _ = env_test.reset()
rewards = []
for i in range(500):
    obs, reward, trunc, term, info = env_test.step(env_test.action_space.sample())
    rewards.append(reward)
    # env_test.render()

print('Mean reward: ', sum(rewards)/len(rewards))
print('Max reward: ', max(rewards))
print('Min reward: ', min(rewards))
print(rewards)
print(sum(rewards))
```
